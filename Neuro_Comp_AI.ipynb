## Hybrid model combining base model with Neuro model

```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
import math
from tqdm import tqdm
import matplotlib.pyplot as plt
from brian2 import *

#########################################
# SynapticAttention Module (Unchanged)
#########################################
class SynapticAttention(nn.Module):
    """
    Attention mechanism inspired by neural synapses with adaptive connection strengths.
    """
    def __init__(self, hidden_size, num_attention_heads, dropout_prob=0.1):
        super().__init__()
        self.num_attention_heads = num_attention_heads
        self.attention_head_size = hidden_size // num_attention_heads
        self.all_head_size = self.num_attention_heads * self.attention_head_size
        
        # Query, Key, Value projections
        self.query = nn.Linear(hidden_size, self.all_head_size)
        self.key = nn.Linear(hidden_size, self.all_head_size)
        self.value = nn.Linear(hidden_size, self.all_head_size)
        
        # Synaptic plasticity parameters (learnable)
        self.synaptic_strength = nn.Parameter(torch.ones(num_attention_heads, 1, 1))
        self.decay_rate = nn.Parameter(torch.ones(num_attention_heads, 1, 1) * 0.9)
        
        # Projection back to hidden size
        self.output = nn.Linear(hidden_size, hidden_size)
        self.dropout = nn.Dropout(dropout_prob)
        
        # Tracking recent activations for plasticity
        self.register_buffer("activation_history", torch.zeros(num_attention_heads, 1, 1))

    def transpose_for_scores(self, x):
        new_x_shape = x.size()[:-1] + (self.num_attention_heads, self.attention_head_size)
        x = x.view(*new_x_shape)
        return x.permute(0, 2, 1, 3)
    
    def forward(self, hidden_states, attention_mask=None):
        # Standard attention calculations
        mixed_query_layer = self.query(hidden_states)
        mixed_key_layer = self.key(hidden_states)
        mixed_value_layer = self.value(hidden_states)
        
        query_layer = self.transpose_for_scores(mixed_query_layer)
        key_layer = self.transpose_for_scores(mixed_key_layer)
        value_layer = self.transpose_for_scores(mixed_value_layer)
        
        # Dot product between query and key to get attention scores
        attention_scores = torch.matmul(query_layer, key_layer.transpose(-1, -2))
        attention_scores = attention_scores / math.sqrt(self.attention_head_size)
        
        if attention_mask is not None:
            attention_scores = attention_scores + attention_mask
        
        # Apply synaptic strength modulation
        modulated_attention_scores = attention_scores * self.synaptic_strength
        
        # Softmax normalization
        attention_probs = F.softmax(modulated_attention_scores, dim=-1)
        attention_probs = self.dropout(attention_probs)
        
        # Update synaptic strengths based on current activations (Hebbian learning)
        activation_intensity = torch.mean(torch.abs(attention_probs), dim=[-2, -1], keepdim=True)
        self.activation_history = self.decay_rate * self.activation_history + (1 - self.decay_rate) * activation_intensity
        
        # Apply attention to value layer
        context_layer = torch.matmul(attention_probs, value_layer)
        context_layer = context_layer.permute(0, 2, 1, 3).contiguous()
        
        new_context_layer_shape = context_layer.size()[:-2] + (self.all_head_size,)
        context_layer = context_layer.view(*new_context_layer_shape)
        
        output = self.output(context_layer)
        
        return output, attention_probs

#########################################
# NeuroSynapticLM for GPTNeoX Architecture
#########################################
class NeuroSynapticLM(nn.Module):
    """
    A small language model enhanced with neural synapse-inspired mechanisms.
    Modified to work with GPTNeoX architecture.
    """
    def __init__(self, base_model_name="EleutherAI/pythia-160m"):
        super().__init__()
        print("Loading base model:", base_model_name)
        # Load a small pretrained language model
        self.tokenizer = AutoTokenizer.from_pretrained(base_model_name)
        self.base_model = AutoModelForCausalLM.from_pretrained(base_model_name)
        
        # Extract configuration
        config = self.base_model.config
        hidden_size = config.hidden_size
        num_heads = config.num_attention_heads
        
        print(f"Model config: hidden_size={hidden_size}, num_heads={num_heads}")
        print(f"Model architecture: {type(self.base_model).__name__}")
        
        # For GPTNeoX, the layers are in gpt_neox.layers instead of transformer.h
        if hasattr(self.base_model, 'gpt_neox'):
            num_layers = len(self.base_model.gpt_neox.layers)
            print(f"Detected GPTNeoX architecture with {num_layers} layers")
            print(f"Replacing attention in layers {num_layers-3} to {num_layers-1} with synaptic attention")
            
            # Replace the last 3 attention layers with synaptic attention
            for i in range(num_layers-3, num_layers):
                original_attn = self.base_model.gpt_neox.layers[i].attention
                synaptic_attn = SynapticAttention(
                    hidden_size=hidden_size,
                    num_attention_heads=num_heads
                )
                
                if hasattr(original_attn, 'query_key_value'):
                    qkv_weight = original_attn.query_key_value.weight.data
                    qkv_size = qkv_weight.size(0) // 3
                    synaptic_attn.query.weight.data.copy_(qkv_weight[:qkv_size])
                    synaptic_attn.key.weight.data.copy_(qkv_weight[qkv_size:2*qkv_size])
                    synaptic_attn.value.weight.data.copy_(qkv_weight[2*qkv_size:])
                    
                    if hasattr(original_attn, 'dense'):
                        synaptic_attn.output.weight.data.copy_(original_attn.dense.weight.data)
                
                class AttentionWrapper(nn.Module):
                    def __init__(self, synaptic_attention):
                        super().__init__()
                        self.synaptic_attention = synaptic_attention
                    
                    def forward(self, hidden_states, attention_mask=None, position_ids=None, layer_past=None, 
                                head_mask=None, use_cache=False, output_attentions=False, cache_position=None, 
                                position_embeddings=None, **kwargs):
                        output, attention_probs = self.synaptic_attention(hidden_states, attention_mask)
                        return output, attention_probs
                
                self.base_model.gpt_neox.layers[i].attention = AttentionWrapper(synaptic_attn)
        else:
            print(f"Unrecognized model architecture: {type(self.base_model).__name__}")
            raise ValueError("Unsupported model architecture")
        
        # Add a short-term memory buffer for context enhancement
        self.context_buffer = nn.Parameter(torch.zeros(1, 10, hidden_size))
        self.context_gate = nn.Linear(hidden_size, 1)
        print("Model initialization complete")
        
    def forward(self, input_ids, attention_mask=None):
        outputs = self.base_model(input_ids, attention_mask=attention_mask, output_hidden_states=True)
        if hasattr(outputs, 'hidden_states') and outputs.hidden_states is not None:
            last_hidden = outputs.hidden_states[-1]
            gate_values = torch.sigmoid(self.context_gate(last_hidden))
            context_contribution = torch.mean(last_hidden, dim=1, keepdim=True)
            self.context_buffer.data = torch.cat([self.context_buffer[:, 1:], context_contribution], dim=1)
        return outputs.logits

#########################################
# Utility: Compute Distinct Metrics
#########################################
def compute_distinct_metrics(text):
    """
    Compute Distinct-1 and Distinct-2 metrics:
    - Distinct-1: ratio of unique unigrams to total tokens.
    - Distinct-2: ratio of unique bigrams to total bigrams.
    """
    tokens = text.split()
    if not tokens:
        return 0.0, 0.0
    distinct1 = len(set(tokens)) / len(tokens)
    bigrams = list(zip(tokens, tokens[1:]))
    distinct2 = len(set(bigrams)) / len(bigrams) if bigrams else 0.0
    return distinct1, distinct2

#########################################
# Comparison Function
#########################################
def run_neural_synapse_comparison():
    print("Starting neural synapse model comparison...")
    
    # Define test prompts
    test_prompts = [
        "The capital of France is",
        "Neural networks function by",
        "The most effective way to solve complex problems is",
        "The relationship between quantum mechanics and general relativity is",
        "Climate change solutions require"
    ]
    
    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")
    
    # Load base model
    print("Loading base model...")
    model_name = "EleutherAI/pythia-160m"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    base_model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
    base_model.eval()
    
    # Create neural synapse enhanced model
    print("Creating NeuroSynapticLM model...")
    neural_model = NeuroSynapticLM(base_model_name=model_name).to(device)
    neural_model.eval()
    
    # Generation parameters
    gen_params = {
        'max_length': 50,
        'do_sample': True,
        'temperature': 0.7,
        'num_return_sequences': 1
    }
    
    results = {}
    
    for prompt in test_prompts:
        print(f"\n\n=== PROMPT: {prompt} ===")
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        
        results[prompt] = {
            'base': {},
            'neural': {}
        }
        
        # Generate with base model
        print("\n--- BASE MODEL OUTPUT ---")
        with torch.no_grad():
            base_outputs = base_model.generate(
                inputs.input_ids, 
                **gen_params
            )
        base_text = tokenizer.decode(base_outputs[0], skip_special_tokens=True)
        print(base_text)
        base_words = base_text.split()
        base_word_count = len(base_words)
        base_unique_words = len(set([word.lower() for word in base_words]))
        base_distinct1, base_distinct2 = compute_distinct_metrics(base_text)
        
        results[prompt]['base'] = {
            'text': base_text,
            'word_count': base_word_count,
            'unique_words': base_unique_words,
            'distinct1': base_distinct1,
            'distinct2': base_distinct2
        }
        
        # Generate with neural synapse model
        print("\n--- NEURAL SYNAPSE MODEL OUTPUT ---")
        with torch.no_grad():
            _ = neural_model(inputs.input_ids)
            neural_outputs = base_model.generate(
                inputs.input_ids,
                **gen_params
            )
        neural_text = tokenizer.decode(neural_outputs[0], skip_special_tokens=True)
        print(neural_text)
        neural_words = neural_text.split()
        neural_word_count = len(neural_words)
        neural_unique_words = len(set([word.lower() for word in neural_words]))
        neural_distinct1, neural_distinct2 = compute_distinct_metrics(neural_text)
        
        results[prompt]['neural'] = {
            'text': neural_text,
            'word_count': neural_word_count,
            'unique_words': neural_unique_words,
            'distinct1': neural_distinct1,
            'distinct2': neural_distinct2
        }
        
        # Compute and print Jaccard similarity between base and neural outputs
        base_set = set([word.lower() for word in base_words])
        neural_set = set([word.lower() for word in neural_words])
        if len(base_set.union(neural_set)) > 0:
            jaccard = len(base_set.intersection(neural_set)) / len(base_set.union(neural_set))
        else:
            jaccard = 0.0
        print("\n--- COMPARISON METRICS ---")
        print(f"Base model    - Words: {base_word_count}, Unique words: {base_unique_words}, Distinct-1: {base_distinct1:.4f}, Distinct-2: {base_distinct2:.4f}")
        print(f"Neural model  - Words: {neural_word_count}, Unique words: {neural_unique_words}, Distinct-1: {neural_distinct1:.4f}, Distinct-2: {neural_distinct2:.4f}")
        print(f"Word overlap (Jaccard similarity): {jaccard:.4f}")
        results[prompt]['jaccard'] = jaccard
    
    # Test for context retention across prompts (outputs printed)
    print("\n\n=== TESTING CONTEXT RETENTION ===")
    context_prompts = [
        "AI systems need to be aligned with human values.",
        "This requires careful engineering and testing.",
        "What are the key principles for ensuring AI safety?"
    ]
    
    print("\nSequential prompts to test context retention:")
    for i, prompt in enumerate(context_prompts):
        print(f"\nPrompt {i+1}: {prompt}")
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        with torch.no_grad():
            base_out = base_model.generate(inputs.input_ids, **gen_params)
        base_out_text = tokenizer.decode(base_out[0], skip_special_tokens=True)
        with torch.no_grad():
            _ = neural_model(inputs.input_ids)
            neural_out = base_model.generate(inputs.input_ids, **gen_params)
        neural_out_text = tokenizer.decode(neural_out[0], skip_special_tokens=True)
        print("\nBase model output:")
        print(base_out_text)
        print("\nNeural model output:")
        print(neural_out_text)
    
    print("\n=== COMPARISON COMPLETE ===")
    return results

#########################################
# Visualization Function
#########################################
def visualize_results(results):
    """
    Visualizes numerical comparisons between base and neural model outputs.
    Generates bar charts for word count, unique word count, distinct metrics, and Jaccard similarity across prompts.
    """
    prompts = list(results.keys())
    base_word_counts = [results[p]['base']['word_count'] for p in prompts]
    neural_word_counts = [results[p]['neural']['word_count'] for p in prompts]
    
    base_unique_words = [results[p]['base']['unique_words'] for p in prompts]
    neural_unique_words = [results[p]['neural']['unique_words'] for p in prompts]
    
    base_distinct1 = [results[p]['base']['distinct1'] for p in prompts]
    neural_distinct1 = [results[p]['neural']['distinct1'] for p in prompts]
    
    base_distinct2 = [results[p]['base']['distinct2'] for p in prompts]
    neural_distinct2 = [results[p]['neural']['distinct2'] for p in prompts]
    
    jaccard_similarities = [results[p]['jaccard'] for p in prompts]
    
    x = np.arange(len(prompts))
    width = 0.35

    fig, axs = plt.subplots(4, 1, figsize=(12, 20))
    
    # Plot Word Count Comparison
    axs[0].bar(x - width/2, base_word_counts, width, label='Base Model')
    axs[0].bar(x + width/2, neural_word_counts, width, label='Neural Model')
    axs[0].set_title("Word Count Comparison")
    axs[0].set_xticks(x)
    axs[0].set_xticklabels(prompts, rotation=45, ha='right')
    axs[0].legend()
    
    # Plot Unique Word Count Comparison
    axs[1].bar(x - width/2, base_unique_words, width, label='Base Model')
    axs[1].bar(x + width/2, neural_unique_words, width, label='Neural Model')
    axs[1].set_title("Unique Word Count Comparison")
    axs[1].set_xticks(x)
    axs[1].set_xticklabels(prompts, rotation=45, ha='right')
    axs[1].legend()
    
    # Plot Distinct Metrics Comparison (Distinct-1 and Distinct-2)
    axs[2].bar(x - width/2, base_distinct1, width, label='Distinct-1 (Base)')
    axs[2].bar(x + width/2, neural_distinct1, width, label='Distinct-1 (Neural)')
    axs[2].set_title("Distinct-1 Metric Comparison")
    axs[2].set_xticks(x)
    axs[2].set_xticklabels(prompts, rotation=45, ha='right')
    axs[2].legend()
    
    axs[3].bar(x - width/2, base_distinct2, width, label='Distinct-2 (Base)')
    axs[3].bar(x + width/2, neural_distinct2, width, label='Distinct-2 (Neural)')
    axs[3].set_title("Distinct-2 Metric Comparison")
    axs[3].set_xticks(x)
    axs[3].set_xticklabels(prompts, rotation=45, ha='right')
    axs[3].legend()
    
    plt.tight_layout()
    plt.show()
    
    # Plot Jaccard Similarity in a separate figure
    plt.figure(figsize=(10, 4))
    plt.bar(x, jaccard_similarities, width, color='purple')
    plt.title("Jaccard Similarity Between Outputs")
    plt.xticks(x, prompts, rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Print numerical summary
    print("Numerical Comparison Summary:")
    for i, p in enumerate(prompts):
        print(f"\nPrompt: {p}")
        print(f"  Base Model: Words = {base_word_counts[i]}, Unique Words = {base_unique_words[i]}, Distinct-1 = {base_distinct1[i]:.4f}, Distinct-2 = {base_distinct2[i]:.4f}")
        print(f"  Neural Model: Words = {neural_word_counts[i]}, Unique Words = {neural_unique_words[i]}, Distinct-1 = {neural_distinct1[i]:.4f}, Distinct-2 = {neural_distinct2[i]:.4f}")
        print(f"  Jaccard Similarity: {jaccard_similarities[i]:.4f}")
    
    # Compute averages across prompts
    avg_base_dist1 = np.mean(base_distinct1)
    avg_neural_dist1 = np.mean(neural_distinct1)
    avg_base_dist2 = np.mean(base_distinct2)
    avg_neural_dist2 = np.mean(neural_distinct2)
    print("\nAverage Distinct Metrics Across Prompts:")
    print(f"  Base Model: Distinct-1 = {avg_base_dist1:.4f}, Distinct-2 = {avg_base_dist2:.4f}")
    print(f"  Neural Model: Distinct-1 = {avg_neural_dist1:.4f}, Distinct-2 = {avg_neural_dist2:.4f}")

#########################################
# Main Execution
#########################################
if __name__ == "__main__":
    results = run_neural_synapse_comparison()
    visualize_results(results)
```
